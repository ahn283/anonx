# LLM with web url

```python
# import libraries
from langchain_community.chat_models import ChatOllama
from openai import OpenAI
import keyring

# llama3.1 model
llama = ChatOllama(model='llama3.1')

def get_answer_llama(prompt):
    response = llama.invoke(prompt)
    return response.content

# chatgpt model
client = OpenAI(api_key=keyring.get_password('openai', 'key_for_windows'))

def get_answer_gpt(prompt, model='gpt-4o'):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {'role':'system',
             'content':'You are a helpful assistant.'},
            {'role':'user',
             'content':prompt}
        ]
    )
    return response.choices[0].message.content
    
```

```python
from newspaper import Article

def get_content_from_url(url):
    article = Article(url)
    article.download()
    article.parse()
    return article.text
    
```

```python
def get_answer(prompt, url, model='llama'):
    content = get_content_from_url(url)
    query = prompt + "\n\n#### article:\n\n" + content
    if model == 'llama':
        answer = get_answer_llama(query)
    else:
        answer = get_answer_gpt(query, model)
    return answer    
```

# Gdeltdoc library

```python
prompt = """You are an information extractor specialized in identifying corporations listed on NASDAQ or Dow Jones. 
Extract key information from the given text and convert it into JSON format. Don't include any other thing but output format.

Follow these steps and respond only in the specified output format:

# Step 1: Identify the main company.

# Step 2: Analyze the sentiment (positive or negative) associated with each company in the context.

# Step 3: Assign a sentiment score between 0 and 1 for each company, with a brief explanation for the score.

# Step 4: Convert the information into JSON format.

#### output format:
{'entity':<company_name>, 'ticker':<stock_ticker>, 'sentiment':<positive/negative>, 'score':<float>, 'datetime':<provider publishing date -> 'YYYY-MM-DD'>}
"""
```

```python
from gdeltdoc import GdeltDoc, Filters
from datetime import datetime
from newspaper import Article
from tqdm.notebook import tqdm
import numpy as np

def get_content_from_url(url):
    article = Article(url)
    article.download()
    article.parse()
    return article.text

def get_answer(prompt, url, model='llama'):
    content = get_content_from_url(url)
    query = prompt + "\n\n#### article:\n\n" + content
    if model == 'llama':
        answer = get_answer_llama(query)
    else:
        answer = get_answer_gpt(query, model)
    return answer, content    


def get_news_url(keyword='Apple', start_date='2020-01-01', end_date='today'):
    if end_date == 'today':
        now = datetime.now().strftime('%Y-%m-%d')
        end_date = now
    f = Filters(
        keyword=keyword,
        start_date=start_date,
        end_date=end_date
    )
    
    gd = GdeltDoc()
    
    links = gd.article_search(f)
    return links
    
    # search for articles matching the filters

def crawl_url(df):
    urls = []
    titles = []
    texts = []
    seen_dates = []
    entities = []
    tickers = []
    scores = []
    publish_dates = []
    for i in tqdm(range(len(df))):
        url = df['url'][i]
        urls.append(url)
        title = df['title'][i]
        titles.append(title)
        seen_date = df['seendate'][i]
        seen_dates.append(seen_date)
        try:
            answer, content = get_answer(prompt, url=url)
            answer = eval(answer)
            text = content
            texts.append(text)
            # print(answer)
            entity = answer['entity']
            entities.append(entity)
            ticker = answer['ticker']
            tickers.append(ticker)
            score = answer['score']
            scores.append(score)
            publish_date = answer['datetime']
            publish_dates.append(publish_date)
        except Exception as e:
            # print(e)
            texts.append('')
            entities.append('')
            tickers.append('')
            scores.append(np.nan)
            publish_dates.append('')
            continue
    articles = pd.DataFrame({
        'title': titles,
        'url': urls,
        'seen_date': seen_dates,
        'texts': texts,
        'entity': entities,
        'ticker': tickers,
        'score': scores,
        'publish_date': publish_dates
    })
        
    return articles
```

```python
links_test = get_news_url(start_date='2024-07-01')
links_test
```

```python
articles = crawl_url(links_test)
articles
```
